```python
import pandas as pd
```

```python
temp_df = pd.read_csv('../src/temp_data.csv')
```

```python
demand_df = pd.read_csv('../src/elec_demand.csv')
```

```python
big_df = demand_df.merge(temp_df, how='inner', left_on='time', right_on='time')
```

```python
big_df.head()
```

```python
big_df.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1, inplace=True)
```

```python
big_df.info()
```

```python
big_df.time.min()
```

```python
big_df.to_csv('../src/big_demand_temp.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=[10,8])
fig, ax1 = plt.subplots(figsize=[10,8])

ax1.set_xlabel('Date')
ax1.set_ylabel('Electricity Demand (MWH)', color='b')
ax1.plot(big_df.loc[:500, 'time'], big_df.loc[:500, 'demand'], c='b')

ax2 = ax1.twinx()

ax2.set_xlabel('Date')
ax2.set_ylabel('Los Angeles Temperature (F)', color='r')
ax2.plot(big_df.loc[:500, 'time'], big_df.loc[:500, 'Los Angeles'], c='r')
```

```python
plt.show()
```
